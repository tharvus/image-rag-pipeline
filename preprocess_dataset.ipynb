```python
# clean csv before saving
df_to_save = new_df[new_df["is_downloaded"] == True]
# change ids
```

```python
# can reset the row ids. 
df_to_save = df_to_save.reset_index()
df_to_save = df_to_save.drop("id", axis=1)
df_to_save.index = range(1, len(df_to_save)+1)
df_to_save["id"] = range(1, len(df_to_save)+1)
```

```python
# now we change up image paths
folder_path = "./images"
img_nums = [int(re.search(r"img_(\d+)\.jpg", f).group(1)) for f in os.listdir(folder_path)]
img_nums.sort()  # sort to maintain order
```

```python
img_nums
```

```python
for idx in range(len(img_nums)):
    img_num_to_rename = img_nums[idx]

    new_name = f"img_{idx+1}.jpg"
    old_name = f"img_{img_num_to_rename}.jpg"

    old_path = os.path.join(folder_path, old_name)
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    
```
